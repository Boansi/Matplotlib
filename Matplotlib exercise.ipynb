```python
from matplotlib import pyplot as plt
import numpy as np 

plt.style.use('classic')

age_x = [23,24,25,26,27,28,29,30]
x_indexes = np.arange(len(age_x))

width = 0.20
salaries_y = [2345,3456,1800,4530,1500,5000,5043,5690]

plt.bar(x_indexes - width, salaries_y, linewidth = 2, width=width,color = 'c', linestyle='-',label ='All Develops')

python_developers = [3009,3689,3567,6783,7783,5683,8763,6538]
plt.bar (x_indexes, python_developers, width=width, color = 'g', label ='Python Develops')

java_script = python_developers = [2684,4783,7663,2743,4345,2424,5545,6445]
plt.bar(x_indexes + width, java_script, width=width, color = 'b', label = 'Javascript')

plt.xlabel('Ages')
plt.ylabel('Salaries in (EUR)')
plt.title('Salary median')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.xticks(ticks=x_indexes, labels=age_x)
plt.savefig('first_graph.png')
plt.show()
```
